```python
def markov_dict():
  # defining a function
    word_freqs = {}
    # setting an empty dictionary for word frequencies
    with open("sonnets.txt", "r") as file_object:
        # reading our input file
        past_word = ""
        # setting past_word to an empty string
        for line in file_object:
            line_split = line.split()
          # looping over each line in our input file
            for curr_word in line_split:
                # splitting each line into individual words
                if past_word in word_freqs:
                  # checking if the past word exists
                    frequency = word_freqs[past_word]
                    # if exists, assign it to our dictionary
                else:
                    frequency = {}
                    # if it doesn't exist add it with a new key to our dictionary
                    word_freqs[past_word] = frequency
                if curr_word in frequency:
                  # if exists increment the value by 1
                    frequency[curr_word] += 1
                else:
                  # if not, assign value to 1
                    frequency[curr_word] = 1
                    
                past_word = curr_word
                # update past_word to the current word we're on
        # del word_freqs[" "]
        return word_freqs
        # return the word frequencies dictionary
```
